* Question: In pyhton there is already list then why we use numpy array ??

Answer: There are three main benefits of numpy array over a python list: 
* (1) Less Memory 
* (2) Fast 
* (3) Convinient

```python
# Less memory
import numpy as np
import sys

print(sys.version) # for getting the version of sys

l = range(1000)
print(sys.getsizeof(1)*len(l))  # take 28 bytes of memory means 1 element = 28 bytes

array = np.arange(1000)
print(array.size * array.itemsize) # 1 element = 4 bytes 
#here array.size gives total length of array and array.itemsize gives the size of single element
```

```python
# Fast 
# We will use two list of numbers using python list and numpy array and compare them 

import time

Size = 1000000

l1 = range(Size)
l2 = range(Size)

a1 = np.arange(Size)
a2 = np.arange(Size)

# python list
start = time.time()
result = [(x+y) for x,y in zip(l1,l2)]
print("Python list took time: ",(time.time()-start)*1000) #by defult time in seconds so we put 1000 here for miliseconds

# numpy array
start = time.time()
result = a1 + a2 #covinient because when we add two array we simply write a1+a2 
print("Numpy took time: ",(time.time()-start)*1000)
```

```python
# Convinient
a1 = np.array([1,2,3])
a2 = np.array([4,5,6])

print(a1+a2)

print(a2-a1)

print(a1*a2)

print(a1/a2)
```

```python
# ndim
import numpy as np
a = np.array([5,6,7])  #one dimensional array
print("1-D Array: \n",a)

b = np.array([[1,2],[3,4],[5,6]])
print("\n\n2-D Array:\n",b)


print(a.ndim)  #it tells the dimension of metrix
print(b.ndim)  
```

```python
# itemsize
import numpy as np
a = np.array([5,6,7])
print(a.itemsize)  #it tells the size of a single element
# There is integer that why the here the size is bytes 

b = np.array([[1,2],[3,4],[5,6]])
print("\n\n2-D Array:\n",b)
print(b.itemsize)
```

```python
# dtype
a = np.array([1,2,3])
print(a.dtype)  # it tells the data type of array
print('Bytes used: ',a.itemsize)

x = np.array([[1,2],[3,4],[5,6]], dtype = np.float64)  # it can change the datatype 
print(x)
print("Bytes used: ",x.itemsize)
x
```

```python
# size
x = np.array([[1,2],[3,4],[5,6]], dtype=np.float64)
print(x)
x.size # total number of element present in array
```

```python
# shape 
x = np.array([[1,2],[3,4],[5,6]], dtype=np.float64)
print(x)
x.shape  # gives the shape of matrix i.e width and height(rows,column)
```

```python
x = np.array([[1,2],[3,4],[5,6]], dtype=complex)
x
```

```python
# Sometime we want to initialize array with zeros then we use zeros
# we simply write np.zeros( (shape) )
np.zeros( (3,4) )
```

```python
# initialize matrix with ones
np.ones( (3,5) )
```

```python
# arange it is similar to range function in list
l=range(5)
print(l)
print(l[0])
print(l[1])
print(l[2])

#similarly 
print(np.arange(1,5)) # it gives array according to given range
print(np.arange(1,5,2)) #steps jump
```

```python
#linspace
print(np.linspace(1,5,10)) # it will generate 10 numbers which is linearly space
print("\n\n", np.linspace(1,5,5))
print("\n\n",np.linspace(1,5,20))
```

```python
# reshape
a = np.array([[1,2],[3,4],[5,6]])
print(a)
print("\n",a.shape)

print("\n\n",a.reshape(2,3))


a.reshape(6,1)
```

```python
#ravel
a = np.array([[1,2],[3,4],[5,6]])
print(a)

print("\n\n",a.ravel()) # it will flatten the array or make it one dimension
print("\n\n",a)  #ravel don't change original array it will just return a new array
```

```python
#min
a = np.array([[1,2],[3,4],[5,6]])
print(a)
print("\n\n",a.min())
```

```python
#max
a = np.array([[1,2],[3,4],[5,6]])
print(a)
print("\n\n",a.max())
```

```python
# axis=0 deal with  column value
a = np.array([[1,2],[3,4],[5,6]])
print(a)
a.sum(axis=0)
```

```python
# axis=1 deal with  each row value
a = np.array([[1,2],[3,4],[5,6]])
print(a)
a.sum(axis=1)
```

```python
#sum
import numpy as np
a = np.array([[1,2],[3,4],[5,6]])
print(a)
a.sum()
```

```python
#sqrt
a = np.array([[1,2],[3,4],[5,6]])
print(a)
np.sqrt(a)
```

```python
#std (standard deviation)
a = np.array([[1,2],[3,4],[5,6]])
print(a)
np.std(a)
```

```python
#Some array operation
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print(a)
print("\n\n",b)
print("\n\n",a+b)
print("\n\n",a-b)
print("\n\n",a*b)
print("\n\n",a/b)
```

```python
#a.dot(b)
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print(a)
print("\n\n",b)
print("\n\n",a.dot(b))  #matrix product
```
