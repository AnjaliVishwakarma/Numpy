#  Iterator numpy array using nditer

```python
import numpy as np
```

```python
a = np.arange(12).reshape(3,4)
a
```

```python
for row in a:
    print(row)
    
```

```python
for row in a:
    for cell in row:
        print(cell)
```

```python
# for row in a:
#     for cell in row:
#         print(cell)
# Or
for cell in a.flatten():
    print(cell)
```

```python
for x in np.nditer(a,order='C'): # here 'C' order means row by row element access 
    print(x)                      # and Fortran order mean column by column
```

```python
for x in np.nditer(a,order='F'):      #Fortran order
    print(x)
```

```python
for x in np.nditer(a,order='F',flags=['external_loop']):  # print entire value column by column
    print(x)
```

```python
for x in np.nditer(a,op_flags=['readwrite']):
    x[...]=x*x  #
print(a)
```

```python
b = np.arange(3,15,4).reshape(3,1)
b
```

```python
# iterate two arrays simultaneously
for x,y in np.nditer([a,b]):
    print(x,y)  # its possible only when they both are equals or one of them is one
```
