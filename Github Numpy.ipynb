```python
# reverse a array in numpy
# Ques: Print the reverse NumPy array with type float.
import numpy

def arrays(arr):
    # complete this function
    # use numpy.array
    arr.reverse()
    numpy_arr = numpy.array(arr,float)
    return numpy_arr

    
arr = input().strip().split(' ')
result = arrays(arr)
print(result)
```
